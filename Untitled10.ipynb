```python
#Task 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes
    
    print("Total Likes:", total_likes)
    
    # Check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes_data = [200, 300, 150, 400]  # Total = 1050
analyze_post_engagement(likes_data)
```

```python
#Task 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total += rain
    
    # Calculate average rainfall
    average = total / len(rainfall_data)
    
    print("Average Rainfall:", int(average))
    
    # Check rainfall status
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
seasonal_rainfall = [70, 75, 68, 80, 67]
required = 70

rainfall_checker(seasonal_rainfall, required)
```

```python
#Task 4: Social Media – Duplicate Account DetectionTask 4: Social Media – Duplicate Account Detection
def check_duplicate_accounts(usernames):
    # Convert list to set
    unique_usernames = set(usernames)
    
    # Compare lengths
    if len(unique_usernames) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
users = ["manisha01", "rahul23", "sneha45", "rahul23"]

check_duplicate_accounts(users)
```

```python
#Task 5: Healthcare – Appointment Eligibility Checker
# Function to check appointment eligibility
def check_eligibility(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

# Taking input from user
age = int(input("Enter Patient Age: "))

# Calling the function
status = check_eligibility(age)

# Printing output in expected format
print("Patient Age:", age)
print("Eligibility Status:", status)
```

```python
#Task 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:   # Premium condition
            premium.append(price)

    return premium


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 1999]

premium_crops = filter_premium_crops(crop_prices)

print("Premium Crops:", premium_crops)
```

```python
#7: System Monitoring – Application Health
def check_application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print("Error Count:", errors)
    print("System Status:", status)

# Example usage
check_application_health(7)
```

```python
#Task 8: Banking – Daily Transaction Limit Checker
def check_transaction(amount):
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    
    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Example usage
check_transaction(60000)
```

```python
#Task 9: E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    present_count = 0

    # Loop to count present days
    for status in attendance_list:
        if status == "P":
            present_count += 1

    # Calculate attendance percentage
    attendance_percentage = (present_count / total_classes) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    # Print output in expected format
    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", eligibility)


# Example attendance record (P = Present, A = Absent)
student_attendance = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]

check_exam_eligibility(student_attendance)
```

```python
#Task 10: Smart Electricity Bill Analyzer
def calculate_bill(units):
    bill = 0

    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    # Classify usage
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units_consumed = int(input("Enter units consumed: "))
total_bill, usage_status = calculate_bill(units_consumed)

print("Total Electricity Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
